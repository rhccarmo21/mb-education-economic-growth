```python
# ==========================================================
# Projeto: Educação e Crescimento Econômico – Notebook Completo
# ==========================================================

# 1️⃣ Bibliotecas
import os
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from statsmodels.nonparametric.smoothers_lowess import lowess
import textwrap

# Configuração de estilo profissional
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 11
plt.rcParams['figure.titlesize'] = 18
plt.rcParams['figure.titleweight'] = 'bold'
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# 2️⃣ Criação das pastas (mantido igual)
os.makedirs('data/raw', exist_ok=True)
os.makedirs('data/processed', exist_ok=True)
os.makedirs('results', exist_ok=True)
os.makedirs('reports', exist_ok=True)

# 3️⃣ Função para coletar dados (mantida igual)
def fetch_worldbank_indicator(indicator_code, countries='all', start_year=2000, end_year=2023):
    per_page = 1000
    page = 1
    records = []
    while True:
        url = f'http://api.worldbank.org/v2/country/{countries}/indicator/{indicator_code}?date={start_year}:{end_year}&format=json&per_page={per_page}&page={page}'
        response = requests.get(url)
        data = response.json()
        if not data or len(data) < 2 or not data[1]: break
        for entry in data[1]:
            records.append({
                'country': entry['country']['value'],
                'country_iso': entry['country']['id'],
                'year': int(entry['date']),
                'value': entry['value']
            })
        page += 1
        if page > data[0]['pages']: break
    return pd.DataFrame(records)

# 4️⃣ Coleta de indicadores (mantido igual)
print("Coletando dados do Banco Mundial...")
df_gdp = fetch_worldbank_indicator('NY.GDP.PCAP.CD')          # PIB per capita
df_edu = fetch_worldbank_indicator('SE.XPD.TOTL.GD.ZS')       # Gasto em Educação (% PIB)
print("Dados coletados.")

# Salvar dados brutos
df_gdp.to_csv('data/raw/worldbank_gdp.csv', index=False)
df_edu.to_csv('data/raw/worldbank_edu.csv', index=False)

# 5️⃣ Merge e padronização (mantido igual)
df_global = pd.merge(df_gdp, df_edu, on=['country','country_iso','year'], how='outer', suffixes=('_gdp','_edu'))

# Mapear continentes e regiões para análise
continent_map = {
    'USA':'América do Norte','BRA':'América Latina','ARG':'América Latina','MEX':'América Latina',
    'DEU':'Europa','FRA':'Europa','GBR':'Europa','RUS':'Europa',
    'CHN':'Ásia','IND':'Ásia','JPN':'Ásia','KOR':'Ásia',
    'ZAF':'África','EGY':'África','NGA':'África',
    'AUS':'Oceania','NZL':'Oceania'
}
df_global['region'] = df_global['country_iso'].map(continent_map).fillna('Outros')

# 6️⃣ Pré-processamento para gráficos (mantido igual)
df_plot = df_global.dropna(subset=['value_gdp','value_edu']).copy()
df_plot['efficiency'] = df_plot['value_gdp'] / df_plot['value_edu']
df_plot_sorted = df_plot.sort_values('value_edu')

# 7️⃣ Cores por região - paleta mais moderna
palette = {
    'América do Norte':'#1f77b4','América Latina':'#ff7f0e','Europa':'#2ca02c',
    'Ásia':'#d62728','África':'#9467bd','Oceania':'#8c564b','Outros':'#7f7f7f'
}

# Função auxiliar para adicionar estatísticas aos gráficos
def add_stats_text(ax, x, y, xpos=0.05, ypos=0.95):
    """Adiciona texto com estatísticas de correlação"""
    corr = np.corrcoef(x, y)[0, 1]
    ax.text(xpos, ypos, f'Correlação: {corr:.3f}', 
            transform=ax.transAxes, fontsize=12,
            bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

# ===========================
# 8️⃣ Scatterplot Educação x PIB per Capita - OTIMIZADO
# ===========================
fig, ax = plt.subplots(figsize=(16, 10))

# Scatterplot com tamanho variável por população (se disponível) ou constante
scatter = ax.scatter(df_plot['value_edu'], df_plot['value_gdp'], 
                    c=df_plot['region'].map(palette), 
                    alpha=0.7, s=60, edgecolor='w', linewidth=0.5)

# Adicionar linha de tendência
z = np.polyfit(df_plot['value_edu'], df_plot['value_gdp'], 1)
p = np.poly1d(z)
ax.plot(df_plot['value_edu'], p(df_plot['value_edu']), "r--", alpha=0.8, linewidth=2, 
        label=f'Tendência (y = {z[0]:.0f}x + {z[1]:.0f})')

# Melhorar labels e título
ax.set_xlabel('Gasto Público em Educação (% do PIB)', fontsize=14, fontweight='bold')
ax.set_ylabel('PIB per Capita (USD)', fontsize=14, fontweight='bold')
ax.set_title('Relação entre Investimento em Educação e PIB per Capita (2000-2023)', 
             fontsize=16, fontweight='bold', pad=20)

# Adicionar grid e melhorar escala
ax.grid(True, linestyle='--', alpha=0.3)
ax.set_yscale('log')  # Escala log para melhor visualização

# Legenda customizada
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', 
                             markerfacecolor=color, markersize=10, label=region)
                  for region, color in palette.items()]
ax.legend(handles=legend_elements, title='Região', 
          bbox_to_anchor=(1.05, 1), loc='upper left')

# Adicionar estatísticas
add_stats_text(ax, df_plot['value_edu'], df_plot['value_gdp'])

# Melhorar formatação de números no eixo y
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000:.0f}K'))

plt.tight_layout()
plt.savefig('results/scatter_gdp_edu_global.png', dpi=300, bbox_inches='tight')
plt.show()

# ===========================
# 9️⃣ Gráfico de barras – Eficiência Regional - OTIMIZADO
# ===========================
region_eff = df_plot.groupby('region')['efficiency'].agg(['mean', 'std', 'count']).sort_values('mean', ascending=False)

fig, ax = plt.subplots(figsize=(14, 8))

# Gráfico de barras com barras de erro
bars = ax.bar(region_eff.index, region_eff['mean'], 
              yerr=region_eff['std']/np.sqrt(region_eff['count']),
              color=[palette.get(region, '#7f7f7f') for region in region_eff.index],
              alpha=0.8, edgecolor='black', linewidth=0.5, capsize=5)

# Adicionar valores nas barras
for i, (idx, row) in enumerate(region_eff.iterrows()):
    ax.text(i, row['mean'] + row['std']/10, f'{row["mean"]:,.0f}', 
            ha='center', va='bottom', fontweight='bold')

# Melhorar labels
ax.set_ylabel('Eficiência (PIB per Capita / % Gasto Educação)', fontsize=14, fontweight='bold')
ax.set_xlabel('Região', fontsize=14, fontweight='bold')
ax.set_title('Eficiência do Investimento em Educação por Região', 
             fontsize=16, fontweight='bold', pad=20)

# Rotacionar labels do eixo x se necessário
plt.xticks(rotation=45, ha='right')
ax.grid(axis='y', linestyle='--', alpha=0.3)

# Formatar eixo y
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000:.0f}K'))

plt.tight_layout()
plt.savefig('results/efficiency_by_region_bar.png', dpi=300, bbox_inches='tight')
plt.show()

# ===========================
# 🔟 Curva de Saturação (LOWESS) - OTIMIZADO
# ===========================
fig, ax = plt.subplots(figsize=(16, 10))

# Scatter plot com cores por região
for region, color in palette.items():
    region_data = df_plot_sorted[df_plot_sorted['region'] == region]
    if not region_data.empty:
        ax.scatter(region_data['value_edu'], region_data['value_gdp'], 
                  color=color, alpha=0.6, s=50, label=region, edgecolor='w', linewidth=0.3)

# Suavização LOWESS
smoothed = lowess(df_plot_sorted['value_gdp'], df_plot_sorted['value_edu'], frac=0.3)
ax.plot(smoothed[:,0], smoothed[:,1], color='black', linewidth=3, 
        label='Tendência Suavizada (LOWESS)', linestyle='-')

# Destacar ponto de saturação (onde a derivada muda significativamente)
dy = np.diff(smoothed[:,1]) / np.diff(smoothed[:,0])
saturation_point = smoothed[np.argmin(dy), 0]
ax.axvline(x=saturation_point, color='red', linestyle='--', alpha=0.8, 
           label=f'Ponto de Saturação (~{saturation_point:.1f}% do PIB)')

# Configurações do gráfico
ax.set_xlabel('Gasto Público em Educação (% do PIB)', fontsize=14, fontweight='bold')
ax.set_ylabel('PIB per Capita (USD)', fontsize=14, fontweight='bold')
ax.set_title('Relação não Linear: Retornos Decrescentes do Investimento em Educação', 
             fontsize=16, fontweight='bold', pad=20)
ax.set_yscale('log')
ax.grid(True, linestyle='--', alpha=0.3)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Adicionar anotação explicativa
ax.annotate('Retornos decrescentes além deste ponto', 
            xy=(saturation_point, smoothed[np.argmin(dy), 1]),
            xytext=(saturation_point+2, smoothed[np.argmin(dy), 1]*0.8),
            arrowprops=dict(arrowstyle='->', color='red'),
            fontsize=11, ha='left')

plt.tight_layout()
plt.savefig('results/return_saturation.png', dpi=300, bbox_inches='tight')
plt.show()

# ===========================
# 1️⃣1️⃣ Séries temporais de países selecionados - OTIMIZADO
# ===========================
selected_countries = ['United States','China','Brazil','Germany','India']
df_sel = df_global[df_global['country'].isin(selected_countries)]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 12))

# Gráfico 1: PIB per capita
for country in selected_countries:
    country_data = df_sel[df_sel['country'] == country]
    ax1.plot(country_data['year'], country_data['value_gdp'], 
             marker='o', markersize=4, linewidth=2.5, label=country)

ax1.set_xlabel('Ano', fontsize=14, fontweight='bold')
ax1.set_ylabel('PIB per Capita (USD)', fontsize=14, fontweight='bold')
ax1.set_title('Evolução do PIB per Capita (2000-2023)', fontsize=16, fontweight='bold')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.3)
ax1.set_yscale('log')
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000:.0f}K'))

# Gráfico 2: Gasto em educação
for country in selected_countries:
    country_data = df_sel[df_sel['country'] == country]
    ax2.plot(country_data['year'], country_data['value_edu'], 
             marker='s', markersize=4, linewidth=2.5, label=country)

ax2.set_xlabel('Ano', fontsize=14, fontweight='bold')
ax2.set_ylabel('Gasto em Educação (% do PIB)', fontsize=14, fontweight='bold')
ax2.set_title('Evolução do Investimento em Educação (2000-2023)', fontsize=16, fontweight='bold')
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.3)

plt.tight_layout()
plt.savefig('results/time_series_selected_countries.png', dpi=300, bbox_inches='tight')
plt.show()

# ===========================
# 1️⃣2️⃣ Heatmap de Correlações - NOVO GRÁFICO
# ===========================
# Calcular correlações por região
correlations = {}
for region in df_plot['region'].unique():
    df_r = df_plot[df_plot['region'] == region][['value_gdp', 'value_edu']].dropna()
    if len(df_r) > 10:  # Apenas regiões com dados suficientes
        correlations[region] = df_r['value_gdp'].corr(df_r['value_edu'])

# Criar heatmap
fig, ax = plt.subplots(figsize=(10, 8))
regions = list(correlations.keys())
corr_values = [correlations[region] for region in regions]

# Criar matriz para heatmap
corr_matrix = np.array(corr_values).reshape(-1, 1)
im = ax.imshow(corr_matrix, cmap='RdYlBu_r', vmin=-1, vmax=1, aspect='auto')

# Adicionar valores e labels
for i, value in enumerate(corr_values):
    ax.text(0, i, f'{value:.3f}', ha='center', va='center', 
            fontweight='bold', fontsize=12,
            color='white' if abs(value) > 0.5 else 'black')

# Configurar eixos
ax.set_xticks([0])
ax.set_xticklabels(['Correlação'])
ax.set_yticks(range(len(regions)))
ax.set_yticklabels(regions)
ax.set_title('Correlação entre Educação e PIB per Capita por Região', 
             fontsize=16, fontweight='bold', pad=20)

# Adicionar barra de cores
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Coeficiente de Correlação', fontsize=12)

plt.tight_layout()
plt.savefig('results/correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

# ===========================
# 1️⃣3️⃣ Dashboard Resumo - NOVO GRÁFICO
# ===========================
fig = plt.figure(figsize=(18, 12))
gs = fig.add_gridspec(2, 2)

# Subplot 1: Scatter plot
ax1 = fig.add_subplot(gs[0, 0])
scatter = ax1.scatter(df_plot['value_edu'], df_plot['value_gdp'], 
                     c=df_plot['region'].map(palette), alpha=0.7, s=40)
ax1.set_xlabel('Gasto em Educação (% PIB)')
ax1.set_ylabel('PIB per Capita (USD)')
ax1.set_yscale('log')
ax1.set_title('Relação Educação vs PIB')

# Subplot 2: Barras de eficiência
ax2 = fig.add_subplot(gs[0, 1])
region_eff = df_plot.groupby('region')['efficiency'].mean().sort_values(ascending=False)
ax2.bar(region_eff.index, region_eff.values, 
        color=[palette.get(region, '#7f7f7f') for region in region_eff.index])
ax2.set_xticklabels(region_eff.index, rotation=45, ha='right')
ax2.set_title('Eficiência por Região')
ax2.set_ylabel('Eficiência (PIB/% Educação)')

# Subplot 3: Série temporal
ax3 = fig.add_subplot(gs[1, :])
for country in selected_countries:
    country_data = df_sel[df_sel['country'] == country]
    ax3.plot(country_data['year'], country_data['value_gdp'], 
             marker='o', label=country, linewidth=2)
ax3.set_xlabel('Ano')
ax3.set_ylabel('PIB per Capita (USD)')
ax3.set_title('Evolução Temporal - Países Selecionados')
ax3.legend()
ax3.set_yscale('log')

plt.suptitle('Dashboard Analítico: Educação e Crescimento Econômico', 
             fontsize=18, fontweight='bold')
plt.tight_layout()
plt.savefig('results/analytical_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()

# Relatório de estatísticas
print("\n" + "="*60)
print("RELATÓRIO ANALÍTICO: EDUCAÇÃO E CRESCIMENTO ECONÔMICO")
print("="*60)

# Correlação global
df_corr_global = df_plot[['value_gdp','value_edu']].dropna()
corr_global = df_corr_global['value_gdp'].corr(df_corr_global['value_edu'])
print(f"\nCorrelação global PIB x Educação: {corr_global:.3f}")

# Correlações por região
print("\nCorrelações por região:")
for region in df_plot['region'].unique():
    df_r = df_plot[df_plot['region']==region][['value_gdp','value_edu']].dropna()
    if len(df_r) > 10:
        corr_r = df_r['value_gdp'].corr(df_r['value_edu'])
        print(f"  {region:20}: {corr_r:.3f}")

# Estatísticas descritivas
print(f"\nEstatísticas descritivas:")
print(f"  Média gasto educação: {df_plot['value_edu'].mean():.2f}% do PIB")
print(f"  Mediana gasto educação: {df_plot['value_edu'].median():.2f}% do PIB")
print(f"  Média PIB per capita: ${df_plot['value_gdp'].mean():,.0f}")
print(f"  Mediana PIB per capita: ${df_plot['value_gdp'].median():,.0f}")

print("\n" + "="*60)
```
